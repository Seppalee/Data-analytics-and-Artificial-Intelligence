1 Use gamedata.json for this task. This file contains information of games sold through Steam. Parse out the following information from the data:
- TOP 3 highest metacritic score. Present results using the following format: *Title* has metacritic score of *Score* (for example)
- Games with discount price being 90 % or more. Present results using the following format: *Title* | discount: *Savings* (for example Metal Gear Solid V: Ground Zeroes | score 90.090090)
- Games having metacritic score higher than steam score. Present results using the following format: *Title* has metacritic score of *MetacriticScore* and steam score of *SteamRatingPercent*

```python
import json
```

```python
with open("data_files/gamedata.json") as data:
    gamedata = json.load(data)
```

```python
top_games = sorted(gamedata, key = lambda i: i["metacriticScore"],reverse=True)
best_three = top_games[:3]

for dictionary in best_three:
    print(dictionary["title"] + " has metacritic score of " + dictionary["metacriticScore"])
```

```python
for dictionary in gamedata:
    if float(dictionary["savings"]) >= 90:
        print(dictionary["title"] + " | discount " + dictionary["savings"])
```

```python
for dictionary in gamedata:
    if int(dictionary["metacriticScore"]) > int(dictionary["steamRatingPercent"]):
        print(dictionary["title"] + " has metacritic score of " + dictionary["metacriticScore"] + " and steam score of " + dictionary["steamRatingPercent"])
```

2 Use earthquakes.csv for this task. This file contains information about earthquakes recorded between 1965 and 2016. Earthquake magnitude value describes how strong the earthquake is. Magnitude information can be categorized like presented in the table below (*Source: http://www.geo.mtu.edu/UPSeis/magnitude.html*).

| Magnitude      | Class | Effects |
|----------------|-------|---------|
| 2.5 or less    | Minor | Usually not felt, but can be recorded by seismograph. |
| 2.5 to 5.4     | Light | Often felt, but only causes minor damage. |
| 5.5 to 6.0     | Moderate | Slight damage to buildings and other structures. |
| 6.1 to 6.9     | Strong | May cause a lot of damage in very populated areas. |
| 7.0 to 7.9     | Major | Major earthquake. Serious damage. |
| 8.0 or greater | Great | Great earthquake. Can totally destroy communities near the epicenter. |

Count how many earthquakes have occurred in each class.

```python
import csv
```

```python
with open("data_files/earthquakes.csv","r") as f:
    earthquakes_csv = csv.reader(f)
    earthquakes_data = []
    for row in earthquakes_csv:
        element = row[8]
        earthquakes_data.append(element)

earthquakes_data.remove('Magnitude')

minor = []
light = []
moderate = []
strong = []
major = []
great = []

for i in earthquakes_data:
    if float(i) <= 2.5:
        minor.append(i)
    elif float(i) >= 2.5 and float(i) <= 5.4:
        light.append(i)
    elif float(i) >= 5.5 and float(i) <= 6.0:
        moderate.append(i)
    elif float(i) >= 6.1 and float(i) <= 6.9:
        strong.append(i)
    elif float(i) >= 7.0 and float(i) <= 7.9:
        major.append(i)
    elif float(i) >= 8.0:
        great.append(i)
        
print("Number of minor earthquakes: " + str(len(minor)))
print("Number of light earthquakes: " + str(len(light)))
print("Number of moderate earthquakes: " + str(len(moderate)))
print("Number of strong earthquakes: " + str(len(strong)))
print("Number of major earthquakes: " + str(len(major)))
print("Number of great earthquakes: " + str(len(great)))
    
```

3 Use netflix_titles.xml for this task. This file contains information about Netflix movies and TV shows. **Important:** Movies have duration presented in minutes while TV shows have duration presented in amount of seasons! Parse out the following information from the data:
- Movies released in 2017
- TV show and movie amount (present both counts in separate lines)
- Movies with a length between 15 and 20 minutes (values 15 and 20 included)

```python
import xml.etree.ElementTree as e
from collections import Counter 
```

```python
tree = e.parse("data_files/netflix_titles.xml")
root = tree.getroot()

titles = []
for i in root[0]:
    titles.append(i.tag)
```

```python
netflix_collection = []

for elem in root:
    netflix_title = []
    for subelem in elem:
        netflix_title.append(subelem.text)
    netflix_collection.append(netflix_title)
```

```python
netflix_titles = []
for lst in netflix_collection:
    netflix_title_obj = {}
    for i in range(len(lst)):
        netflix_title_obj[titles[i]] = lst[i]
    netflix_titles.append(netflix_title_obj)
```

```python
released_in_2017 = []
for i in netflix_titles:
    if ("2017" in i["release_year"]):
        released_in_2017.append(i)
        
print(released_in_2017)
```

```python
c1 = Counter()
c2 = Counter()
for item in netflix_titles:
    if item["type"] == "Movie":
        c1[item["type"]] += 1
    elif item ["type"] == "TV Show":
        c2[item["type"]] += 1
    
print(c1)
print(c2)
```

```python
movies = []
for dic in netflix_titles:
    splitted = dic["duration"].split(" ")
    if splitted[1] == "Season":
        continue
    elif splitted[1] == "Seasons":
        continue
    else:
        if int(splitted[0]) >= 15 and int(splitted[0]) <= 20:
            movies.append(dic)

print(movies)
```

4 Use the following Rest API for this task: https://tie.digitraffic.fi/api/v1/data/weather-data. Calculate the average for air temperature (ILMA) and humidity (ILMAN_KOSTEUS) values using two decimals.

```python
import requests
import json
from statistics import mean
```

```python
req = requests.get("https://tie.digitraffic.fi/api/v1/data/weather-data")
weather_data = json.loads(req.text)
```

```python
sensor_values1 = []
sensor_values2 = []
for dic in weather_data["weatherStations"]:
    for sensor in dic["sensorValues"]:
        if sensor["name"] == "ILMA":
            sensor_values1.append(sensor["sensorValue"])
        elif sensor["name"] == "ILMAN_KOSTEUS":
            sensor_values2.append(sensor["sensorValue"])
        else:
            continue

average1 = mean(sensor_values1)
average2 = mean(sensor_values2)

print("Average air temperature: ", round(average1, 2))
print("Average humidity: ", round(average2, 2))        
```

5 Use the following Rest API for this task: https://open-api.myhelsinki.fi/v1/events/. This API provides the information for events held in Helsinki city area. List all music event names in finnish (tip: event tag must have string "music" in it). **Important:** Each event name should only be listed once!

```python
req2 = requests.get("https://open-api.myhelsinki.fi/v1/events/")
event_data = json.loads(req2.text)
```

```python
music_events = []

for dic in event_data["data"]:
    for i in dic["tags"]:
        if i["name"] == "music":
            music_events.append(dic["name"]["fi"])

print(music_events)
```
